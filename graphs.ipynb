```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0,1,2,3,4])
y = np.array([0,1,2,3,4])

plt.plot(x,y, marker = "*")
plt.title("Title")
plt.xlabel("x-axis")
plt.ylabel("y-table")

plt.grid()
plt.show()

```

```python
x_axis = np.array([0,3,5,7])
y_axis = np.array([0,5,8,16])

plt.scatter(x_axis,y_axis)
plt.show()
```

```python
x = np.array(["Apples","Bannas","Mangoes","Grapes"])
y = np.array([4,15,20,2])

plt.bar(x,y, color = "cyan",width=0.4)
plt.show()
```

```python
x = np.array(["Apples","Bannas","Mangoes","Grapes"])
y = np.array([4,15,20,2])

plt.barh(x,y, color = "cyan",height=0.5)
plt.show()
```

```python
x = np.array([70,50,10,90])
my_label = ["Jeep","BMW","Toyota","Volvo"]
my_colors = ["azure","cyan","red","black"]
my_explode = [0.3,0,0,0]
plt.pie(x, labels = my_label, colors = my_colors, explode = my_explode , shadow=True)
plt.title("Cars")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random arrays
x = np.arange(0, 10, 1)
y1 = np.random.randint(0, 10, size=10)
y2 = np.random.randint(0, 10, size=10)
labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']

# Print random arrays
print("x = ", x)
print("y1 = ", y1)
print("y2 = ", y2)
print("labels = ", labels)

# Line graph
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Graph')
plt.legend()
plt.show()

# Bar graph
plt.bar(x, y1, label='Bar 1')
plt.bar(x, y2, bottom=y1, label='Bar 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bar Graph')
plt.legend()
plt.show()

# Pie chart
plt.pie(y1, labels=labels)
plt.title('Pie Chart')
plt.show()
```
